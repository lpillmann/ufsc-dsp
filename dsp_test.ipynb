```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal, misc

import pylab

# References:
# http://docs.scipy.org/doc/scipy/reference/tutorial/signal.html
# http://stackoverflow.com/questions/13145218/stem-plot-in-matplotlib
# 
```

```python
image = misc.face(gray=True)
w = np.zeros((50, 50))
w[0][0] = 1.0
w[49][2] = 1.0
w[49][25] = 1.0
image_new = signal.fftconvolve(image, w)
```

```python
plt.figure()
plt.imshow(image)
plt.gray()
plt.title('Original image')
plt.show()
```

```python
plt.figure()
plt.imshow(image_new)
plt.gray()
plt.title('Filtered image')
plt.show()
```

```python
image = misc.ascent()
w = signal.gaussian(50, 10.0)
image_new = signal.sepfir2d(image, w, w)
```

```python
plt.figure()
plt.imshow(image)
plt.gray()
plt.title('Original image')
plt.show()
```

```python
plt.figure()
plt.imshow(image_new)
plt.gray()
plt.title('Filtered image')
plt.show()
```
