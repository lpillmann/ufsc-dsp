```python
# Import a lot of functions to keep it simple to use
from pylab import * 
```

```python
# parameters
fs = 1000   # [Hz]
Ts = 1.0/fs # [s]
```

```python
# 1 second of data samples at spacing of 1/1000 seconds
t = arange(0, 1, Ts)
#t = arange(0, 1.001, Ts)

# sine wave of 100 Hz
s1 = sin(2*pi*100*t)

# plot first 20 points of the resulting data (why not all 1000?)
plot(s1[:30])
show()
```

```python
# Note: you may need to use fft.fft is you are using ipython
ft = fft(s1)#/len(s1)
plot(20*log10(abs(ft)))
show()
```

```python
# Another signal with two sine waves
s2 = sin(2*pi*100*t)+sin(2*pi*200*t)
plot(s2[:30])
show()
```

```python
# Note: you may need to use fft.fft is you are using ipython
ft = fft(s2)/len(s2)
plot(20*log10(abs(ft)))
show()
```

```python
# adds some noise
noise_amp = 2.0
s_pure = sin(2*pi*100*t) + sin(2*pi*200*t)

s = s_pure + noise_amp*randn(len(t))

plot(s_pure[:30])
plot(s[:30])
show()
```

```python
ft_pure = fft(s_pure)/len(s_pure)
ft = fft(s)/len(s)

plot(20*log10(abs(ft_pure)))
plot(20*log10(abs(ft)))
show()
```

```python
# filter design Remez algorithm
from scipy.signal import remez
lpf = remez(21, [0, 0.2, 0.3, 0.5], [1.0, 0.0])

from scipy.signal import freqz
w, h = freqz(lpf)

plot(w/(2*pi), 20*log10(abs(h))) # w/(2*pi) normalizes X-axis, resulting in 0.5 --> pi
show()
```

```python
# filter design FIR window algorithm
from scipy.signal import firwin
lpf = firwin(numtaps=40, cutoff=0.4, window='hamming') # cutoff is a fraction of Nyquist frequency

from scipy.signal import freqz
w, h = freqz(lpf)

plot(w/(2*pi), 20*log10(abs(h)))
show()
```

```python
# filter design FIR window 2 algorithm
from scipy.signal import firwin2
lpf = firwin2(150, [0.0, 0.3, 0.6, 1.0], [1.0, 2.0, 0.5, 0.0], window='hamming') # cutoff is a fraction of Nyquist frequency

from scipy.signal import freqz
w, h = freqz(lpf)

plot(w/(2*pi), 20*log10(abs(h)))
show()
```

```python
# filtering the signal
from scipy.signal import lfilter

sout = lfilter(lpf, 1, s)

plot(s[:30])
plot(s_pure[:30])
plot(sout[:30])
show()
```

```python
ft = fft(s)/len(s)
ftout = fft(sout)/len(sout)

plot(20*log10(abs(ft)))
plot(20*log10(abs(ftout)))
show()
```
